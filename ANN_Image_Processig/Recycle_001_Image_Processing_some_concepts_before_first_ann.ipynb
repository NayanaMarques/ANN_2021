```python
import zipfile
file_name = zipfile.ZipFile('trash.zip', 'r')
file_name.extractall()
```

```python
# all images are load in variable x...
FILE='trash.zip'
img_name=[]
x_gray=[]
x_original = [] 
with zipfile.ZipFile(FILE, "r") as f:
    for name in f.namelist():
        img_name.append(name)
        x_gray.append(cv2.imread(name, 0)) #leitura da imagem - 0: leitura em preto e branco
        x_original.append(cv2.imread(name))
        #x.append(cv2.resize(cv2.imread(name),(64,64))) #padronizar tamnho da imagem
```

```python
y_names = []
for yy in img_name:
  name = yy[0]+yy[1]+yy[2]
  y_names.append(name)
```

```python
unique_list = []    
for yy in y_names:
  if yy not in unique_list:
    unique_list.append(yy)
```

```python
n = len(unique_list) #incluir na lista todos os possíveis nomes, sem repetições, assim sabemos quais nomes existem
y = []
for yy in y_names:
  for i in range(n):
    if(yy == unique_list[i]):
      y.append(i)
```
